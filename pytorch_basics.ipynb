# PyTorch Basics: NumPy vs PyTorch
This notebook is organized into small sections that compare NumPy and PyTorch for common tensor and matrix operations.
We will cover:
- matrix transpose
- dot product
- matrix multiplication
- softmax
- logarithms and cross-entropy intuition

```python
# NumPy is used for arrays and mathematical operations
import numpy as np

# torch is the main PyTorch package for tensor operations
import torch

# torch.nn contains neural network layers and functions such as Softmax
import torch.nn as nn

# Set a random seed so random values stay the same each time you run the notebook
np.random.seed(7)
torch.manual_seed(7)
```

## 1. Matrix Transpose
The transpose swaps rows and columns of a matrix.
If the original matrix has shape `(2, 5)`, its transpose will have shape `(5, 2)`.

```python
# Create a 2 x 5 matrix using NumPy
numpy_matrix = np.asarray([[1, 0, 2, 5, -2], [1, 0, 2, 5, -2]])

# .shape shows the size as (rows, columns)
print('Original NumPy shape:', numpy_matrix.shape)

# .T means transpose, so rows become columns and columns become rows
numpy_transpose = numpy_matrix.T
print('Transposed NumPy shape:', numpy_transpose.shape)

# Print the new matrix so we can see the row/column swap
print(numpy_transpose)
```

```python
# Create the same 2 x 5 matrix using a PyTorch tensor
torch_matrix = torch.tensor([[1, 0, 2, 5, -2], [1, 0, 2, 5, -2]])

# PyTorch tensors also have a shape
print('Original PyTorch shape:', torch_matrix.shape)

# .T does the transpose in PyTorch too
torch_transpose = torch_matrix.T
print('Transposed PyTorch shape:', torch_transpose.shape)

# Print the transposed tensor
print(torch_transpose)
```

## 2. Dot Product
The dot product multiplies matching elements and adds them together.
It is defined for vectors with the same length.

```python
# Create two 1D NumPy vectors of the same length
nv1 = np.array([1, 2, 3, 4])
nv2 = np.array([1, 2, 3, 4])

# np.dot multiplies matching positions and then adds the results
print('np.dot:', np.dot(nv1, nv2))

# This line shows the same idea in two steps: multiply first, then sum
print('elementwise multiply + sum:', np.sum(nv1 * nv2))
```

```python
# Create two 1D PyTorch tensors of the same length
tv1 = torch.tensor([1, 2, 3, 4])
tv2 = torch.tensor([1, 2, 3, 4])

# torch.dot computes the dot product directly
print('torch.dot:', torch.dot(tv1, tv2))

# This is the same calculation written as multiply first, then sum
print('elementwise multiply + sum:', torch.sum(tv1 * tv2))
```

## 3. Matrix Multiplication
For matrix multiplication, the number of columns in the first matrix must match the number of rows in the second matrix.
Here `(3, 2) x (2, 3)` produces a `(3, 3)` result.

```python
# Create a 3 x 2 matrix with random values
m1 = np.random.rand(3, 2)

# Create a 2 x 3 matrix with random values
m2 = np.random.rand(2, 3)

# Matrix multiplication is valid because the inner sizes match: 2 and 2
numpy_result = np.matmul(m1, m2)

# Show both input matrices
print('Matrix 1:\n', m1)
print('Matrix 2:\n', m2)

# Show the multiplication result and its shape
print('NumPy matmul result:\n', numpy_result)
print('Result shape:', numpy_result.shape)
```

```python
# Create a 3 x 2 PyTorch matrix with random values
m1_torch = torch.randn(3, 2)

# Create a 2 x 3 PyTorch matrix with random values
m2_torch = torch.randn(2, 3)

# torch.matmul performs matrix multiplication
torch_result = torch.matmul(m1_torch, m2_torch)

# Show both input tensors
print('Matrix 1:\n', m1_torch)
print('Matrix 2:\n', m2_torch)

# Show the output tensor and its final shape
print('PyTorch matmul result:\n', torch_result)
print('Result shape:', torch_result.shape)
```

## 4. Softmax Function
Softmax converts raw scores into probabilities that add up to 1.
It is commonly used in classification tasks where one input must be mapped to one of several categories.

```python
# These are raw scores, sometimes called logits
z = np.array([1, 2, 3], dtype=np.float64)

# Softmax formula for each value is: softmax(z_i) = exp(z_i) / sum(exp(z))
# Step 1: compute exp(z_i) for each element
num = np.exp(z)
print('Exponentials of z:', num)

# Step 2: compute the denominator sum(exp(z))
den = np.sum(num)

# Step 3: divide each exp(z_i) by the denominator
softmax_numpy = num / den

# The result is a probability vector, so all values add up to 1
print('Numerator:', num)
print('Denominator:', den)
print('Softmax probabilities:', softmax_numpy)
print('Sum of probabilities:', np.sum(softmax_numpy))
```

```python
# Create the same raw scores as a PyTorch tensor
z_torch = torch.tensor([1.0, 2.0, 3.0])

# Softmax formula: softmax(z_i) = exp(z_i) / sum(exp(z))
# dim=0 means apply the formula across this 1D tensor
softfunc = nn.Softmax(dim=0)
print('Softmax function object:', softfunc)

# This applies the full softmax formula to z_torch
sigma_torch = softfunc(z_torch)

# The output values are probabilities, so their total is 1
print('Softmax probabilities:', sigma_torch)
print('Sum of probabilities:', torch.sum(sigma_torch))
```

## 5. Logarithms, Entropy, and Cross-Entropy
The logarithm is commonly used in loss functions because it turns probabilities into a scale that penalizes wrong confident predictions strongly.
Entropy measures uncertainty in a probability distribution. Cross-entropy compares predicted probabilities with the target distribution.

```python
# Create a small NumPy array
x = np.array([1, 2, 3], dtype=np.float64)

# Log formula: log([x1, x2, x3]) = [log(x1), log(x2), log(x3)]
# Here np.log means the natural logarithm, written as ln(x) in math
print('Values:', x)
print('Natural log of x:', np.log(x))
```

```python
# A probability distribution for three classes
probabilities = np.array([0.7, 0.2, 0.1], dtype=np.float64)

# One-hot target means the first class is the correct answer
target = np.array([1.0, 0.0, 0.0], dtype=np.float64)

# Entropy formula: H(p) = -sum(p * log(p))
# This measures how uncertain the probability distribution is
entropy = -np.sum(probabilities * np.log(probabilities))

# Cross-entropy formula: CE(y, p) = -sum(y * log(p))
# Here y is the target and p is the predicted probability distribution
cross_entropy = -np.sum(target * np.log(probabilities))

# If the target is one-hot like [1, 0, 0], cross-entropy becomes -log(p_correct_class)
print('Probabilities:', probabilities)
print('Entropy:', entropy)
print('Cross-entropy for the target class:', cross_entropy)
```

```python
# Logarithms using PyTorch
torch_x = torch.tensor([1.0, 2.0, 3.0])

# torch.log computes the natural logarithm, written as ln(x) in math
# Formula: log([x1, x2, x3]) = [log(x1), log(x2), log(x3)]
print('Values:', torch_x)
print('Natural log using torch.log:', torch.log(torch_x))
```

## 6. Key Takeaways
- A vector is a 1D tensor.
- A matrix is a 2D tensor.
- Higher-dimensional tensors are used throughout deep learning for data, weights, biases, and intermediate computations.
- PyTorch mirrors many NumPy operations, which makes it easier to learn if you already know NumPy.

```python
# Create two 1D tensors
x1 = torch.tensor([1, 2, 3])
x2 = torch.tensor([10, 20, 30])

# For 1D tensors, matmul behaves like a dot product
# It multiplies matching positions and adds the results
print('Dot product using torch.matmul:', torch.matmul(x1, x2))
```
